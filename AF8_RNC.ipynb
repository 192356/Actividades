```python
# cargamos las librerias necesarias
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos MNIST
(train_images, train_labels), (test_images, test_labels) = keras.datasets.mnist.load_data()

# Mostrar información sobre los datos
print("Forma de las imágenes de entrenamiento:", train_images.shape)
print("Forma de las etiquetas de entrenamiento:", train_labels.shape)
print("Forma de las imágenes de prueba:", test_images.shape)
print("Forma de las etiquetas de prueba:", test_labels.shape)
```

```python
# Normalizar los valores de píxeles al rango [0, 1]
train_images = train_images.astype("float32") / 255.0
test_images = test_images.astype("float32") / 255.0

# Redimensionar las imágenes para añadir el canal (las imágenes son en escala de grises, por lo que solo tienen 1 canal)
train_images = np.expand_dims(train_images, -1)
test_images = np.expand_dims(test_images, -1)

# Convertir las etiquetas a one-hot encoding
num_classes = 10
train_labels = keras.utils.to_categorical(train_labels, num_classes)
test_labels = keras.utils.to_categorical(test_labels, num_classes)

# Mostrar las nuevas formas
print("Nueva forma de las imágenes de entrenamiento:", train_images.shape)
print("Nueva forma de las imágenes de prueba:", test_images.shape)
```

```python
# Definir la arquitectura de la red neuronal convolucional
model = keras.Sequential([
    layers.Conv2D(32, kernel_size=(3, 3), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation="softmax")
])

# Compilar el modelo
model.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

# Mostrar un resumen del modelo
model.summary()

# Entrenar el modelo
batch_size = 128
epochs = 15
history = model.fit(
    train_images, train_labels,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=0.1
)

# Graficar la precisión y pérdida durante el entrenamiento
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Precisión entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión validación')
plt.title('Precisión del modelo')
plt.ylabel('Precisión')
plt.xlabel('Época')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Pérdida entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida validación')
plt.title('Pérdida del modelo')
plt.ylabel('Pérdida')
plt.xlabel('Época')
plt.legend()

plt.show()
```

```python
# Evaluar el modelo con los datos de prueba
score = model.evaluate(test_images, test_labels, verbose=0)
print("Pérdida en los datos de prueba:", score[0])
print("Precisión en los datos de prueba:", score[1])

# Realizar 5 predicciones aleatorias
predictions = model.predict(test_images)
random_indices = np.random.choice(len(test_images), 5, replace=False)

plt.figure(figsize=(15, 3))
for i, idx in enumerate(random_indices):
    plt.subplot(1, 5, i+1)
    plt.imshow(test_images[idx].reshape(28, 28), cmap='gray')
    plt.title(f"Pred: {np.argmax(predictions[idx])}\nReal: {np.argmax(test_labels[idx])}")
    plt.axis('off')
plt.show()
```
