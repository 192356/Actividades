```python
# Importar librerías
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Cargar el dataset desde Excel
from google.colab import files
uploaded = files.upload()

# Leer el archivo Excel
df = pd.read_excel('dataset_sensores_motores.xlsx')

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()

print("Primeras 5 filas del dataset después de limpiar nombres de columnas:")
print(df.head())
print("\nColumnas en el dataset después de limpiar:", df.columns.tolist())
```

```python
# Verificar nombres de columnas
print("Columnas en el dataset:", df.columns.tolist())

# Separar características (X) y etiquetas (y) - Ajusta según lo que imprimió arriba
# Usar nombres de columnas
X = df[['SENSOR1', 'SENSOR2', 'SENSOR3']]  # Datos de entrada
y = df[['MOTOR_A', 'MOTOR_B']]             # Datos de salida

# Proceder con el entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\nDatos de entrenamiento: {len(X_train)} muestras")
print(f"Datos de prueba: {len(X_test)} muestras")
```

```python
# Crear el modelo de Árbol de Decisión
model = DecisionTreeClassifier(max_depth=3, random_state=42)  # Evita sobreajuste

# Entrenar el modelo
model.fit(X_train, y_train)

print("Modelo de Árbol de Decisión entrenado.")

# Realizar predicciones en el conjunto de entrenamiento
y_train_pred = model.predict(X_train)

# Precisión en datos de entrenamiento (evaluar cada salida por separado)
train_accuracy_motor_A = accuracy_score(y_train['MOTOR_A'], y_train_pred[:, 0])
train_accuracy_motor_B = accuracy_score(y_train['MOTOR_B'], y_train_pred[:, 1])

print(f"\nPrecisión en entrenamiento (MOTOR_A): {train_accuracy_motor_A * 100:.2f}%")
print(f"Precisión en entrenamiento (MOTOR_B): {train_accuracy_motor_B * 100:.2f}%")
```

```python
# Realizar predicciones en todo el dataset (X)
y_full_pred = model.predict(X)

# Crear un DataFrame para comparar las predicciones con los valores reales
comparison_full_dataset = pd.DataFrame({
    'Real MOTOR_A': y['MOTOR_A'],
    'Predicho MOTOR_A': y_full_pred[:, 0],
    'Real MOTOR_B': y['MOTOR_B'],
    'Predicho MOTOR_B': y_full_pred[:, 1]
})

print("Comparación de valores reales vs. predichos para todo el dataset:")
display(comparison_full_dataset.head()) # Mostrar las primeras filas para no sobrecargar el output
```

```python
# Obtener reglas del árbol en texto
tree_rules = export_text(model, feature_names=['SENSOR1', 'SENSOR2', 'SENSOR3'])
print("\nReglas del Árbol de Decisión:")
print(tree_rules)
```
